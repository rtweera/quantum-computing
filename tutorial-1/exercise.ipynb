```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
```

```python
qc = QuantumCircuit(5)

qc.h(0)
qc.x(1)
qc.h(2)
qc.y(4)

qc.cx(0, 1)
qc.cx(1, 2)

qc.y(0)
qc.h(1)
qc.y(2)
qc.x(3)
qc.h(4)

qc.cx(2, 3)
qc.cx(2, 4)
qc.cx(3, 4)

qc.h(0)
qc.x(4)

qc.cx(4, 0)

qc.draw("mpl")
```

```python
qc = QuantumCircuit(5)

qc.h(0)
qc.x(1)
qc.h(2)
qc.y(4)
qc.barrier() # just a visualize barrier for better organizing

qc.cx(0, 1)
qc.cx(1, 2)
qc.barrier()

qc.y(0)
qc.h(1)
qc.y(2)
qc.x(3)
qc.h(4)
qc.barrier()

qc.cx(2, 3)
qc.cx(2, 4)
qc.cx(3, 4)
qc.barrier()

qc.h(0)
qc.x(4)
qc.barrier()

qc.cx(4, 0)
qc.barrier()

qc.draw("mpl")
```
